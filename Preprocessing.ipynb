```
# Opens the Video file
import cv2
import os
import glob

train = 1
val = 1
test = 1
count =1
for file_name in glob.glob('*.mp4'):
  file_name_without_ext = os.path.splitext(file_name)[0]
  cap= cv2.VideoCapture(file_name)
  print(file_name,cap.get(cv2.CAP_PROP_FPS))
  i=0
  j =1
  while(cap.isOpened()):
      ret, frame = cap.read()
      if ret == False:
          break
      if(i%25 ==0):
        frame = cv2.resize(frame,(256,256))
        if(j%9==0):
          cv2.imwrite('/content/drive/MyDrive/Research/Colorization/Coloring-greyscale-images/Full-version/Dataset/Test/'+file_name_without_ext+"_"+str(test)+'.jpg',frame)
          test +=1
        elif (j%10 ==0):
          cv2.imwrite('/content/drive/MyDrive/Research/Colorization/Coloring-greyscale-images/Full-version/Dataset/Validation/'+file_name_without_ext+"_"+str(val)+'.jpg',frame)
          val +=1
          j = 0
        else:
          cv2.imwrite('/content/drive/MyDrive/Research/Colorization/Coloring-greyscale-images/Full-version/Dataset/Train/'+file_name_without_ext+"_"+str(train)+'.jpg',frame)
          train +=1
        j+=1
      i+=1
  
  cap.release()
  cv2.destroyAllWindows()

  if count == 11: 
    break
  count+=1

print("training image: ",train)

print("validation image: ",val)

print("test image: ",test)
```
